## Results Presentation - functions

```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
FIGURE_SIZE = (14, 5)
```

```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def h_sigmoid(x):
    return np.where(x > 3, 1, np.where(x < -3, 0, (x + 3) / 6))
```

```python
def swish(x, beta=1):
    return x * (1 / (1 + np.exp(-beta * x)))

def h_swish(x):
    return x * np.where(x > 3, 1, np.where(x < -3, 0, (x + 3) / 6))
```

```python
def plot_sigmoid_hsigmoid_comparision(figure_size: tuple):
    x_values = np.linspace(-8, 8, 500)
    sigmoid_values = sigmoid(x_values)
    h_sigmoid_values = h_sigmoid(x_values)

    custom_labels = ['sigmoid', 'h-sigmoid']
    ylabel = "f(x)"
    colors = plt.cm.viridis(np.linspace(0, 1, 2))

    plt.figure(figsize=figure_size)
    sns.set(style="whitegrid")

    plt.plot(x_values, sigmoid_values, label=custom_labels[0], color=colors[0], linewidth=2)
    plt.plot(x_values, h_sigmoid_values, label=custom_labels[1], color=colors[1], linewidth=2)

    plt.xlabel('x', fontsize=14, fontweight='bold')
    plt.ylabel(ylabel, fontsize=14, fontweight='bold')
    plt.legend(fontsize=12)
    plt.show()
```

```python
def plot_swish_hswish_comparison(figure_size: tuple):
    x_values = np.linspace(-8, 8, 500)
    swish_values = swish(x_values)
    h_swish_values = h_swish(x_values)

    custom_labels = ['swish', 'h-swish']
    ylabel = "f(x)"
    colors = plt.cm.viridis(np.linspace(0, 1, 2))

    plt.figure(figsize=figure_size)
    sns.set(style="whitegrid")

    plt.plot(x_values, swish_values, label=custom_labels[0], color=colors[0], linewidth=2)
    plt.plot(x_values, h_swish_values, label=custom_labels[1], color=colors[1], linewidth=2)

    plt.xlabel('x', fontsize=14, fontweight='bold')
    plt.ylabel(ylabel, fontsize=14, fontweight='bold')
    plt.legend(fontsize=12)
    plt.show()
```

```python
plot_sigmoid_hsigmoid_comparision(figure_size=FIGURE_SIZE)
```

```python
plot_swish_hswish_comparison(figure_size=FIGURE_SIZE)
```
